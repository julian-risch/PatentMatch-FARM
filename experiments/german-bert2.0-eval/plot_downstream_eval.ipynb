```python
import pandas as pd
```

```python
df = pd.read_csv("runs.csv")
```

```python
df['steps_k'] = df["Name"].str.split("_").apply(lambda x: int(x[1].replace("step",""))/1000)
df['task'] = df["Name"].str.split("_").apply(lambda x: x[0])
```

```python
grouped = df.groupby(by=("steps_k", "task")).mean()
```

```python
ge18_coarse = df.loc[df["task"] == "germEval18Coarse"]
ge18_fine = df.loc[df["task"] == "germEval18Fine"]
ge14 = df.loc[df["task"] == "germEval14"]
```

```python
def plot(task_df, y="Test_f1_macro_text_classification"):
    task_df = task_df.sort_values(by="steps_k")
    grouped = task_df.groupby(by="steps_k").mean().reset_index()
    grouped.plot(kind="line", x="steps_k", y=y)
    
```

```python
plot(ge18_coarse, "Test_f1_macro_text_classification")
plot(ge18_fine, "Test_f1_macro_text_classification")
plot(ge14, "Test_seq_f1_ner")
```
